```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow # Importa para exibir imagens no Colab
import numpy as np

# ==============================================================================
# MUITO IMPORTANTE: AJUSTE ESTE CAMINHO!
# O modelo treinado é salvo em um diretório dentro de 'runs/detect/'.
# O nome da pasta de treinamento pode ser 'train', 'train2', 'train3', etc.
# Verifique o painel de arquivos do Colab (à esquerda) na pasta 'runs/detect/'
# para encontrar o nome correto da pasta do seu treinamento mais recente.
# ==============================================================================
trained_model_path = '/content/runs/detect/train3/weights/best.pt' # AJUSTE 'train' E 'best.pt' SE NECESSÁRIO!
model = YOLO(trained_model_path)
```

```python
# --- Escolha UMA das opções abaixo para obter sua imagem de teste ---

# Opção 1: Fazer upload de uma imagem diretamente para o Colab para teste
# from google.colab import files
# print("Por favor, faça o upload de uma imagem de fruta para teste:")
# uploaded = files.upload()
# for filename in uploaded.keys():
#     test_image_path = filename
# -------------------------------------------------------------------

# Opção 2: Usar uma imagem de teste do seu Google Drive (se já estiver montado)
# Se seu Google Drive não estiver montado, execute antes:
# from google.colab import drive
# drive.mount('/content/drive')
# test_image_path = '/content/drive/MyDrive/minhas_imagens_teste/minha_maça_de_teste.jpg' # AJUSTE ESTE CAMINHO
# -------------------------------------------------------------------

# Opção 3 (Mais simples para testar): Baixar uma imagem de fruta da internet
# Esta é uma boa maneira de testar rapidamente.
print("Baixando uma imagem de teste da internet...")
!wget -O test_image_fruit.jpg https://upload.wikimedia.org/wikipedia/commons/thumb/4/4c/Bananas.jpg/1200px-Bananas.jpg
test_image_path = '/content/test_image_fruit.jpg' # Corrected path to the downloaded image
print(f"Imagem de teste baixada para: {test_image_path}")
# -------------------------------------------------------------------


# Realiza a inferência (detecção de objetos)
# O modelo vai processar a imagem e encontrar as frutas.
print(f"\nRealizando detecção na imagem: {test_image_path}")
results = model(test_image_path)

# Processa e exibe os resultados
for r in results:
    # 'r.plot()' desenha as caixas delimitadoras (bounding boxes) e os rótulos
    # diretamente sobre a imagem. Esta é a função de visualização principal!
    im_array = r.plot()

    # O OpenCV (cv2) usa um formato de cor BGR, mas para exibir corretamente no Colab
    # com `cv2_imshow`, é melhor converter para RGB.
    im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)

    # Exibe a imagem com as detecções.
    cv2_imshow(im_rgb)

    # Opcional: Imprimir os detalhes das detecções em formato de texto
    print("\nDetalhes das Detecções Encontradas:")
    if len(r.boxes) == 0:
        print("  Nenhuma fruta detectada.")
    else:
        for box in r.boxes:
            class_id = int(box.cls[0]) # ID da classe detectada
            class_name = model.names[class_id] # Nome da classe (ex: 'apple', 'banana')
            confidence = box.conf[0] # Confiança da detecção (0 a 1)
            # Coordenadas da caixa delimitadora [x1, y1, x2, y2]
            xyxy = box.xyxy[0].cpu().numpy().astype(int)
            print(f"  Classe: {class_name}, Confiança: {confidence:.2f}, Bounding Box: {xyxy}")
```
